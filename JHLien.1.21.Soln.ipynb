1.21 The thermal conductivity in a particular plane wall depends as follows on the wall temperature: k = A+ BT, where A and B are constants. The temperatures are TI and T2 on either side if the wall, and its thickness is L. Develop an expression for $q''$. 

In this problem we write the 1D steady heat equation

$$q''=-k(T){dT\over dx}=-(A+BT){dT\over dx}$$

We integrate this expression from x=0 to x=L.  Make sure to plot the result.

$$qL=- AT-B/2 T^2 \left. \right|_{T_1}^{T_2}$$

$$qL=A{T_1}+B/2 {T_1}^2-(A{T_2}+B/2 {T_2}^2)$$

Make up values for A=1, B=0.01, T1=300., T2=330., L=0.025


```python
A=1.;B=0.01; T1=300.;T2=630.; L=0.025;
```

```python
q=(A*T1+.5*B*T1**2 - (A*T2 + 0.5*B* T2**2))/L; q
```

We find the temperature distribution to be:

$$B/2 {T(x)}^2 + AT(x)+qx-A{T_1}-B/2 {T_1}^2=0$$

$$T(x)^2 + {2A \over B}T(x)+{2\over B}(qx-A{T_1}-B/2 {T_1}^2)=0$$

Define 

$a=1$, $b={2A \over B}$, and $c={2\over B}(qx-A{T_1}-B/2 {T_1}^2)$.

$$T(x)={-b + \sqrt{b^2 -4ac} \over 2a}$$
